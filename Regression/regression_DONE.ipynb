```python
import pandas as pd

#берем тренировочные данные
df = pd.read_csv('train.csv')
df.head() 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = df['excerpt'].values
vectorizer = CountVectorizer() #векторизация (эмбеддинг) текста
bag = vectorizer.fit_transform(corpus) #Fit all transformers, transform the data and concatenate results.
#fit используется только на обучающих (train) данных, потому что этот метод изучает данные и запоминвает внутренние параметры
new_data = pd.DataFrame(bag.toarray(), columns=vectorizer.get_feature_names_out()) #Get output feature names for transformation.
new_data['Target'] = df['target']
new_data.head(10)
```

```python
from sklearn.model_selection import train_test_split #эта функция помогает разделить данные на тренировочные и тестовые
X_train, X_test, y_train, y_test = train_test_split(new_data.drop(columns=['Target']), new_data.Target, test_size=0.25, random_state=42)
# для тестирования модели оставьте четверть от имеющихся данных
# выведите по 6 первых строк образованных объектов
X_train.head(6)
```

```python
# создаем и обучаем модель линейной регрессии
from sklearn.linear_model import LinearRegression

modelLR = LinearRegression() #создается объект модели линейной регрессии
modelLR.fit(X_train, y_train) # процесс обучения модели: подбираются такие коэффициенты, которые позволят
#минимизировать ошибку предсказания на трен.данных X_train (признаки) и y_train (целевая переменная)

#прогнозируем значения
y_train_pred = modelLR.predict(X_train) #модель используется для предсказания значений целевой переменной
#обучаться модель должна только на тренировочных данных
```

```python
# спрогнозируйте значения целевой переменной для тестовых объектов
y_test_pred = modelLR.predict(X_test)
```

```python
# импортируем методы для оценки качества модели (метрики качества предсказаний)
from sklearn.metrics import root_mean_squared_error, r2_score, mean_absolute_error, max_error
rmse_train = root_mean_squared_error(y_train, y_train_pred) #среднеквадратичная ошибка: насколько в среднем модель ошибается на предсказаниях
r2_train = r2_score(y_train, y_train_pred) #коээфициент детерминации r^2: показывает, насколько хорошо модель объясняет разброс значений
mae_train = mean_absolute_error(y_train, y_train_pred)
me_train = max_error(y_train, y_train_pred)

rmse_test = root_mean_squared_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)
me_test = max_error(y_test, y_test_pred)


# посмотрите, какие еще метрики могут быть использованы для оценки регрессионной модели:
# https://scikit-learn.org/1.5/api/sklearn.metrics.html
# добавьте новый показатель в итоговую таблицу
# добавьте в итоговую таблицу те же самые показатели для тестовых данных
```

R2 score -- это метрика, показывающая, насколько хорошо модель объясняет разброс целевой переменной. Принимает значения от - (минус) бесконечности до 1. Иначе говоря, эта метрика показывает, насколько хорошо модель подходит к данным: чем ближе к 1, тем лучше. Также называется коэффициентом детерминации.

```python
#импорт и обучение модели с помощью алгоритма случайный лес
from sklearn.ensemble import RandomForestRegressor
#этот алгоритм строит мн-во decision trees и объединяет их результаты для предсказания. В регрессионной модели предсказания всех деревьев усредняются.
modelRF = RandomForestRegressor(max_depth=2)
modelRF.fit(X_train, y_train)

y_train_rf_pred = modelRF.predict(X_train) #The predicted regression target of an input sample is computed as the mean predicted regression targets of the trees in the forest.
y_test_rf_pred = modelRF.predict(X_test)

rmse_rf_train = root_mean_squared_error(y_train, y_train_rf_pred)
r2_rf_train = r2_score(y_train, y_train_rf_pred)
mae_rf_train = mean_absolute_error(y_train, y_train_rf_pred)
me_rf_train = max_error(y_train, y_train_rf_pred)

rmse_rf_test = root_mean_squared_error(y_test, y_test_rf_pred)
r2_rf_test = r2_score(y_test, y_test_rf_pred)
mae_rf_test = mean_absolute_error(y_test, y_test_rf_pred)
me_rf_test = max_error(y_test, y_test_rf_pred)



# добавьте новую запись с результатами оценки текущей модели в конец таблицы.
# кол-во столбцов и показатели - должны быть такие же как настроили на предыдущем этапе
```

```python
results = pd.DataFrame(columns=[
    'Method',
    'R2',
    'Root Mean Squared Error',
    'Mean Absolute Error',
    'Max Error'])
```

```python
results.loc[len(results)] = ['Linear Regression (Train)', r2_train, rmse_train, mae_train, me_train]
results.loc[len(results)] = ['Linear Regression (Test)', r2_test, rmse_test, mae_test, me_test]
results.loc[len(results)] = ['Random forest (Train)', r2_rf_train, rmse_rf_train, mae_rf_train, me_rf_train]
results.loc[len(results)] = ['Random forest (Test)', r2_rf_test, rmse_rf_test, mae_rf_test, me_rf_test]
results
```

```python
# добавьте еще 1-2 модели в этот эксперимент
# https://scikit-learn.org/stable/api/index.html

from sklearn.linear_model import Ridge
modelR = Ridge()

modelR.fit(X_train, y_train)
y_ridge_train_pred = modelR.predict(X_train)
y_ridge_test_pred = modelR.predict(X_test)

rmse_ridge_train = root_mean_squared_error(y_train, y_ridge_train_pred)
r2_ridge_train = r2_score(y_train, y_ridge_train_pred)
mae_ridge_train = mean_absolute_error(y_train, y_ridge_train_pred)
me_ridge_train = max_error(y_train, y_ridge_train_pred)

rmse_ridge_test = root_mean_squared_error(y_test, y_ridge_test_pred)
r2_ridge_test = r2_score(y_test, y_ridge_test_pred)
mae_ridge_test = mean_absolute_error(y_test, y_ridge_test_pred)
me_ridge_test = max_error(y_test, y_ridge_test_pred)

results.loc[len(results)] = ['Ridge (Train)', rmse_ridge_train, r2_ridge_train, mae_ridge_train, me_ridge_train]
results.loc[len(results)] = ['Ridge (Test)', rmse_ridge_test, r2_ridge_test, mae_ridge_test, me_ridge_test]
results
```

```python
from sklearn.neighbors import KNeighborsRegressor
modelKN = KNeighborsRegressor(n_neighbors=2)

modelKN.fit(X_train, y_train)
y_kn_train_pred = modelKN.predict(X_train)
y_kn_test_pred = modelKN.predict(X_test)

rmse_kn_train = root_mean_squared_error(y_train, y_kn_train_pred)
r2_kn_train = r2_score(y_train, y_kn_train_pred)
mae_kn_train = mean_absolute_error(y_train, y_kn_train_pred)
me_kn_train = max_error(y_train, y_kn_train_pred)

rmse_kn_test = root_mean_squared_error(y_test, y_kn_test_pred)
r2_kn_test = r2_score(y_test, y_kn_test_pred)
mae_kn_test = mean_absolute_error(y_test, y_kn_test_pred)
me_kn_test = max_error(y_test, y_kn_test_pred)

results.loc[len(results)] = ['KNeighborsRegression (Train)', rmse_kn_train, r2_kn_train, mae_kn_train, me_kn_train]
results.loc[len(results)] = ['KNeighborsRegression (Test)', rmse_kn_test, r2_kn_test, mae_kn_test, me_kn_test]
results
```

```python
# загрузите новые данные (называется test.csv, но это в контексте kaggle -для проверки моделей участников.
# Для нас это просто новые данные, для которых нужно предсказать значение целевого атрибута.)
# выполните векторизацию текстов и предскажите для них значение признака target с помощью выбранной модели.
df3 = pd.read_csv('train.csv')

vectorizer = CountVectorizer()
bag2 = vectorizer.fit_transform(df3['excerpt'])  

X = pd.DataFrame(bag2.toarray(), columns=vectorizer.get_feature_names_out())
y = df3['target']

modelR = Ridge()
modelR.fit(X, y)
```

```python
# Загружаем новые данные
df_test = pd.read_csv('test.csv')

# Используем vectorizer, обученный выше
bag_test = vectorizer.transform(df_test['excerpt']) 
df_test['pred_target'] = modelR.predict(bag_test)

df_test[['id', 'pred_target']].to_csv('results.csv', index=False)
```
